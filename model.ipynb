Steps to pre process
1. Find a standard training dictionary of 10,000 words plus probably 100-500 medical words or possibly just the medical words
2. Convert each sentence into a vector 

```python
y = [0 for i in range(0, len(hc)+len(nhc))]
x = hc + nhc
for i in range(0, len(hc)):
    y[i] = 1
    
def vectorize(x):
    vector = [0 for i in range(0, len(words))]
    for i in range(0, len(words)):
        if (words[i] in x) or (words[i]+"s" in x) or (words[i]+"es" in x) or (words[i]+"ed" in x) or (words[i]+"ing" in x):
            vector[i] = 1
    return vector

x_vectorize = [vectorize(x[i]) for i in range(0, len(x))]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_vectorize, y, test_size=0.33, random_state=42)
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(len(words),)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

X_train_true = X_train[:int(len(X_train)/1.3)]
X_train_val = X_train[int(len(X_train)/1.3):]
y_train_true = y_train[:int(len(y_train)/1.3)]
y_train_val = y_train[int(len(y_train)/1.3):]


history = model.fit(X_train_true,
                    y_train_true,
                    epochs=10,
                    batch_size=32,
                    validation_data=(X_train_val, y_train_val))
```

```python
history_dict = history.history
history_dict.keys()
import matplotlib.pyplot as plt
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.clf()
acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']
plt.plot(epochs, acc_values, 'bo', label='Training accuracy')
plt.plot(epochs, val_acc_values, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, y_test)
```
